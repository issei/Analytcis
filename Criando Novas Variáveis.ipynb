## Criando Novas Variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

Criaremos uma nova variável chamada Valor Bruto, que consistirá na união dos valores das variáveis Valor, Condominio e IPTU.

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

Em seguida, criaremos outra variável que armazenará o valor por metro quadrado, chamaremos essa nova variável de Valor m2. Iremos coletar Valor e dividi-lo por Area.

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
```

```python
dados.head(10)
```

A formatação pode ficar pouco elegante devido à quantidade de casas decimais. Para resolvermos esse problema utilizaremos o método round(), que em tradução livre significa "arredondar":

```python
dados['Valor m2'] = dados['Valor m2'].round(2)
```

```python
dados.head(10)
```

Nosso próximo passo é criar uma variável para o valor do bruto do metro quadrado(Valor Bruto m2), então, escreveremos:

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.head(10)
```

Agora, criaremos uma variável que abrigue os tipos de imóvel casa e apartamento, para tanto, devemos coletar a variável Tipo e recolher esses marcadores e indentifica-los. Criaremos uma lista com a identificação das casas, então:

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

Assim feito, criaremos efetivamente a nova variável que chamaremos de Tipo Agregado. Depois, evocaremos dados e passaremos a variável Tipo, seguido do método apply(). Esse método pertmite que apliquemos uma função à cada registro do DataFrame. Para tanto, criaremos uma função lambda : 'lambda x: 'Casa' if x in casa else 'Apartamento'.

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

A nova variável foi criada com sucesso! Contudo, nós sabemos que a variável IPTU possui alguns dados faltantes, então Valor Bruto pode armazenar dados imprecisos. Precisamos verificar com a seguradora se manteremos essa estrutura.

```python
dados
```

## Excluindo Variáveis

Chamaremos nosso banco de dados auxiliar de dados_aux, que receberá partes do do banco dados. Em seguida, passaremos uma lista de variáveis que serão utilizadas: Tipo Agregado, Valor m2, Valor Bruto e Valor Bruto m2.

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
```

```python
dados_aux.head(10)
```

Para excluirmos uma variável do banco de dados, usaremos o termo del, passaremos o DataFrame em questão e explicitaremos o nome da variável a ser deletada:

```python
del dados_aux['Valor Bruto']
```

```python
dados_aux.head(10)
```

Outra maneira de excluir variáveis é utilizar o método pop(). Excluiremos neste exemplo a variável Valor Bruto m2:

```python
dados_aux.pop('Valor Bruto m2')
```

```python
dados_aux
```

Uma maneira de excluirmos múltiplas variáveis é utilizando o método drop(). Dessa vez, utilizaremos nosso DataFrame original, realmente deletando as variáveis Valor Bruto e Valor Bruto m2. No caso do método drop(), precisaremos especificar o eixo em que se dará a exclusão, uma vez que ele pode atuar especificamente em linhas ou colunas. Então, escreveremos axis = 1, afinal o eixo 0 corresponde às linhas e o 1 colunas. Precisaremos, ainda, assinalar inplace como True.

```python
dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True)
```

```python
dados.head(10)
```

Finalmente, precisaremos exportar e sobrescrever o arquivo do banco de dados, de forma que essa seja sua versão mais recente e atualizada. Para tanto, escreveremos:

```python
dados.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```

O primeiro deles é *isnull()*. Tal método irá gerar um DataFrame booleano, em que a observação marcada como **True caracteriza um dado nulo**, isto é, um missing.

```python
dados.isnull()
```

Temos, ainda, o método notnull(), que funciona exatamente da maneira inversa ao isnull(): se a informação for nula, será utilizada a notação False.

Contudo, ainda que marquemos true ou false nos dados, ainda não temos a capacidade de analisá-los integralmente. Aprendemos logo nas primeiras aulas alguns usos do método info(), que nos exibe informações do DataFrame.

## Tratamento de Dados Faltantes (continuação)

verificaremos quantas assinaturas nulas temos em Condominio:

```python
dados[dados['Condominio'].isnull()].shape[0]
```

Agora precisamos filtrar os dados de acordo com o tipo Apartamento.

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

Contudo, se nos atentarmos a nossa seleção, verificaremos que ela nos fornece exatamente as assinaturas que deverão ser removidas do banco de dados. Reaproveitaremos a estrutura já utilizada anteriormente neste notebook, e passaremos a seleção que criamos (selecao). Para eliminar as assinaturas nulas selecionadas ao invés de coletá-las, iremos adicionar ~ em selecao, que inverte a Series booleana.

```python
A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape[0]
A - B
```

Foram eliminadas 745 elementos da nossa base de dados. Vamos verificar quantas assinaturas nulas temos em Condominio agora:

```python
dados[dados['Condominio'].isnull()].shape[0]
```

Teremos o total de 1813 assinaturas nulas. Agora precisamos filtrar os dados de acordo do tipo Apartamento com Condomino nulo.

```python
dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
```

o resultado será 1068. O que faremos é manter esses dados, mas atribuir o valor 0 a eles. Temos uma função para esta ação fillna(), que receberá o valor 0 e inplance = True.

```python
dados[dados['Condominio'].isnull()].shape[0]
```

Outra maneira de resolver a questão, é criar um dicionário que indicará os valores das variáveis. Podemos, inclusive, inserir valores diferentes para ambas as variáveis.

```python
dados[dados['IPTU'].isnull()].shape[0]
```

Ao verificarmos novamente a quantidade de nulos existentes em nosso DataFrame, tanto para variável Condominio quanto para IPTU, o resultado será 0. Podemos verificar com mais clareza escrevendo:

```python
dados.info()
```

Temos, portanto, 21826 registros e nenhum deles é nulo. Para finalizarmos essa etapa, devemos salvar este arquivo, de forma que o DataFrame seja atualizado com as modificações que realizamos. Para tanto, escreveremos - lembrando que não queremos incluir o índice:

```python
dados.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```
